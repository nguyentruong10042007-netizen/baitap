```python
c=str(input("nhap vao 2 ki tu:"))
if len(c) !=1:
    print("khoi dong lai code va chi nhap vao 1 ki tu")
else:
    if 'a' <= c <= 'z':
        print("ki tu hoa tuong ung la:", c.upper())
    if 'A' <= c <= 'Z':
        print("ki tu thuong tuong ung la:", c.lower())
```

```python
kt=str(input("nhap vao 1 ki tu:"))
if len(kt) !=1:
    print("chay lai code va nhap 1 ki tu")
else:
    if 'a' <= kt <= 'z' or 'A' <= kt <= 'Z':
        print("{", kt,"}", "la ki tu alphabet")
    else:
        print("{",kt,"}","khong phai ki tu alphabet")
```

```python
import math
a, b, c = map(float, input("nhap vao 3 so:").split())
if a+b>c and b+c>a and a+c>b:
    p=(a+b+c)/2
    s=math.sqrt(p*(p-a)*(p-b)*(p-c))
    print(s)
else:
    print("khong phai 3 canh cua tam giac")
```
