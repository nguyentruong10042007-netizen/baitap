```python
a=int(input("nhap so:"))
b=int(input("nhâp so:"))
c=int(input("nhap so:"))
denta=b**2-4*a*c
if denta>0:
  x1=(-b+sqrt(denta))/2*a
  x2=(-b-sqrt(denta))/2*a
  print(x1, x2)
elif denta==0:
  x1=x2=-b/2*a
  print(x1, x2)
else:
  print("Phuong trinh vo nghiem")
```
